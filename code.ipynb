```python
infographics['whatsapp'] = np.nan
infographics['facebook'] = np.nan
infographics['snapchat'] = np.nan
infographics['twitter'] = np.nan
infographics['instagram'] = np.nan
infographics['reddit'] = np.nan
```

```python
def update_socials():
    def update_platforms(row):
        return int('Reddit' in row), int('WhatsApp' in row), int('Twitter' in row), int('Facebook' in row), int('Instagram' in row), int('Snapchat' in row)

    infographics['reddit'], infographics['whatsapp'], infographics['twitter'], infographics['facebook'], infographics['instagram'], infographics['snapchat'] = zip(*infographics['platforms'].apply(update_platforms))

update_socials()
```

```python
infographics.platforms = infographics.platforms.astype(str)
```

```python
# Regions defined according to the World Bank
regions = {
    'Brazil': 'Latin America',
    'Chile': 'Latin America',
    'China': 'Asia and Pacific',
    'Egypt': 'MENA',
    'France': 'Europe',
    'Germany': 'Europe',
    'Greece': 'Europe',
    'Hungary': 'Europe',
    'Iran': 'MENA',
    'Italy': 'Europe',
    'Lithuania': 'Europe',
    'New Zealand': 'Asia and Pacific',
    'Palestine': 'MENA',
    'Poland': 'Europe',
    'Portugal': 'Europe',
    'Romania': 'Europe',
    'Saudi Arabia': 'MENA',
    'Spain': 'Europe',
    'Syria': 'MENA',
    'Taiwan': 'Asia and Pacific',
    'The Netherlands': 'Europe',
    'UK': 'Europe',
    'USA': 'North America',
    'Ukraine': 'Europe',
    'Venezuela': 'Latin America',
    'Vietnam': 'Asia and Pacific'
}
```

```python
def assign_region(country):
    return regions.get(country)
```

```python
infographics['region'] = infographics['nationality'].apply(assign_region)
```
